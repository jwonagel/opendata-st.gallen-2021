```python
df_hour = df.groupby('timestamp', as_index=False).sum()[['timestamp', 'Additional Energy Export', 'Leistung in kWp', 'Fläche in m2']]
df_hour.head()
```

```python
df['date'] = df['timestamp'].map(lambda x: x.replace(second=0, microsecond=0, minute=0, hour=0, tzinfo=None))
```

```python
df_day = df.groupby('date', as_index=False).sum()[['date', 'Additional Energy Export', 'Leistung in kWp', 'Fläche in m2']]
df_day['Leistung in kWp'] = df.groupby('date', as_index=False).mean()[['Leistung in kWp']]
df_day.describe()
```

```python
df_weather['date'] = df_weather['date'].map(lambda x: datetime.strptime(str(x), '%Y%m%d'))
df_weather.head()
```

```python
df_weather_rel = df_weather[['date', 'tre200d0', 'sre000d0', 'rre150d0']]
```

```python
train_test_data = pd.merge(df_day, df_weather_rel, on='date', how='inner')
train_test_data.head()
```

```python
train_test_data = train_test_data.rename(columns={'tre200d0': 'temp', 'sre000d0': 'sunshine', 'rre150d0': 'rain'})
train_test_data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
Y = train_test_data['Additional Energy Export']
X = train_test_data.drop(['Additional Energy Export'], axis=1)
X['dayofyear'] = X['date'].map(lambda x: x.timetuple().tm_yday) 
X = X.drop(['date'], axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, reg.predict(X_test))
```

```python
y_pred = reg.predict(X_test)
```

```python
weatherforecast = pd.read_csv('./weatherforecast.csv')
weatherforecast.head()
```

```python

weatherforecast['datetime'] = weatherforecast['datetime'].map(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
weatherforecast.head(1)
```

```python
weatherforecast['date'] = weatherforecast['datetime'].map(lambda x: x.replace(second=0, microsecond=0, minute=0, hour=0, tzinfo=None))
weatherforecast['dayofyear'] = weatherforecast['date'].map(lambda x: x.timetuple().tm_yday) 

weatherdaily = weatherforecast[['dayofyear', 'rainfall', 'sunshine']].groupby('dayofyear', as_index=False).sum()
weatherdaily['temperature'] = weatherforecast[['dayofyear', 'temperature']].groupby('dayofyear', as_index=False).mean()['temperature']
weatherdaily.head()
```
