```python
import pandas as pd
from datetime import datetime
import dateutil.parser
```

```python
path_photovoltiac = './../../raw-data/weather/stromproduktion-der-solaranlagen-der-stgaller-stadtwerke.csv'
df = pd.read_csv(path_photovoltiac, sep=';')
```

```python
df.describe()
```

```python
df
```

```python
df['timestamp'] = df['DateTime (Local Time)'].map(lambda x: dateutil.parser.isoparse(x))
```

```python
df.head(100)
```

```python
df['timestamp'] = df['timestamp'].map(lambda x: x.replace(second=0, microsecond=0, minute=0))
```

```python
df.groupby('Name').count()
```

```python
df_hour = df.groupby('timestamp', as_index=False).sum()[['timestamp', 'Additional Energy Export', 'Leistung in kWp', 'Fläche in m2']]
df_hour.head()
```

```python
df['timestamp_day'] = df['timestamp'].map(lambda x: x.replace(second=0, microsecond=0, minute=0, hour=0, tzinfo=None))
```

```python
df_day = df.groupby('timestamp_day', as_index=False).sum()[['timestamp_day', 'Additional Energy Export', 'Leistung in kWp', 'Fläche in m2']]
df_day.head()
```
